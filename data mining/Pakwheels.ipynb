```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def extracter(start,step,data,no_of_values):
    result=[]
    for i in range(len(data)//no_of_values):
        result.append(data[start])
        start+=step
    return result
```

```python
def get_data(page_no,url):
    
    # initialization
    all_data = []
    
    url = url+str(page_no)
    
    response = requests.get(url)
    html_content = response.content
    soup = BeautifulSoup(html_content,'html.parser')
    print(soup.prettify())
    
    # LOOP
    for d in soup.findAll('div', attrs={'class':'well search-list clearfix ad-container page-'+str(page_no)}):
        test = []
        mergedList =[]
        
        # Make & Model
        modelMake = d.find('div', attrs={'class':'search-title'})
        modelMake = modelMake.h3
        test.append(modelMake.text.strip())
        
        # Registered
        city = d.find('ul', attrs={'class':'list-unstyled search-vehicle-info fs13'})
        city = city.find('li')
        test.append(city.text.strip())
        
        # Price
        price = d.find('div', attrs={'class':'price-details generic-dark-grey'})
        test.append(price.text.strip())
        
        # Year, Km, Fuel, CC, Mode
        others = d.find('ul', attrs={'class':'list-unstyled search-vehicle-info-2 fs13'})
        for value in others.find_all('li'):
            mergedList.append(value.get_text())
       
        year = extracter(0,5,mergedList,5)
        for i in year:
            test.append(i.strip())
            
        km = extracter(1,5,mergedList,5)
        for j in km:
            test.append(j.strip())
            
        fuel = extracter(2,5,mergedList,5)
        for k in fuel:
            test.append(k.strip())
            
        cc = extracter(3,5,mergedList,5)
        for l in cc:
            test.append(l.strip())
            
        mode = extracter(4,5,mergedList,5)
        for m in mode:
            test.append(m)
        
        all_data.append(test)
        
    return all_data
```

```python
suzuku_url = 'https://www.pakwheels.com/used-cars/suzuki/32?page='
toyota_url = 'https://www.pakwheels.com/used-cars/toyota/33?page='
honda_url = 'https://www.pakwheels.com/used-cars/honda/14?page='
```

```python
results = []

for i in range(1,2): #1200
    results.append(get_data(i,suzuku_url))
for i in range(1,2): #1500
    results.append(get_data(i,toyota_url))
# for i in range(1,3): #800
#     results.append(get_data(i,honda_url))
    
print(results)
```

```python
flatten = lambda l: [item for sublist in l for item in sublist] 
df = pd.DataFrame(flatten(results),columns=['Model','city','prices','year', 'km', 'type','cc','mode'])
df.to_csv('crack.csv', index=False, encoding='utf-8')
```
